```python
from csv import writer

cores = [1, 2, 4, 8, 10, 14, 16, 20]
sizes = [100, 200, 400, 600, 800, 1000, 1500, 2000]

for i in range(len(cores)):
    for j in range(len(sizes)):
        df = pd.read_csv('./Soluciones-Ubuntu-2/MM1c-Size' + str(sizes[j]) + '-core' + str(cores[i]) + '.csv', header = None, names = ['core', 'time'])

        df_2 = df.groupby('core', as_index=False)['time'].mean()
        df_2['time'] = df_2['time'].round(2)

        with open('times.csv', 'a') as f_object:
        
            # Pass this file object to csv.writer()
            # and get a writer object
            writer_object = writer(f_object)
        
            # Pass the list as an argument into
            # the writerow()
            for ind in df_2.index:
                writer_object.writerow([cores[i], sizes[j], df_2['core'][ind], df_2['time'][ind]])
        
            # Close the file object
            f_object.close()
```

```python
df = pd.read_csv('./times.csv')
df.to_csv('./times.csv', index=False)
```
